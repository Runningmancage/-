```python
from PIL import Image
import os

ori_path = r'dataset_license/train/images'
#token_list = os.listdir(ori_path)
data_path = r'dataset_license/train/images_train'
```

```python
if not os.path.exists(data_path):
    os.mkdir(data_path)
    
```

```python
data_list = os.listdir(ori_path)
print(len(data_list))
```

```python
print(data_list[0])
```

```python
import cv2
import numpy as np

for name in data_list:
    im = cv2.imread(ori_path + '/' + rf'{name}')
    height, width, channel = im.shape
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    image_blur = cv2.GaussianBlur(gray, ksize=(5,5), sigmaX=0)#이미지를 잘 인식시키기위해 가우시안 블러 사용

    kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]]) #커널 생성
    image_sharp = cv2.filter2D(image_blur,-1,kernel) # 블러 처리 된 이미지를 선명화 시키는 작업
#선명화된 이미지를 이진화 처리(영역마다 빛 조건이 달라질 때 도움이 됨) -> 번호판은 빛 조건이 달라 질 수 있어서 이진화를 함
#cv2.adaptiveThreshold(입력 이미지, 최댓값, 적응형 이진화 플래그, 임곗값 형식, 블록 크기, 감산값)
    max_output_value = 255 #출력 픽셀 강도의 최대값을 결정
    neighborhood_size = 51 #블록의 크기(홀수만 가능)
    substract_from_mean = 10 #감산값
    image_binarized = cv2.adaptiveThreshold(image_sharp, max_output_value,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY,neighborhood_size,substract_from_mean)
    contours, _ = cv2.findContours(
    image_binarized,
    mode=cv2.RETR_LIST,
    method=cv2.CHAIN_APPROX_SIMPLE
    )

    temp_result = np.zeros((height, width, channel), dtype=np.uint8)

    cv2.drawContours(temp_result, contours=contours, contourIdx=-1, color=(255,255,255))
    cv2.imwrite(data_path + '/' + name,temp_result)
    
```
